### 기본 개념
- Pandas의 시리즈나 데이터프레임에 plot이라는 메서드가 내장되어있다.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 한글 지언
mpl.rcParams['font.family'] ='batang'
```

```python
np.random.seed(0)

df = pd.DataFrame(np.random.randn(100,3),
                 index = pd.date_range('2018-1-1', periods = 100),
                 columns = list('ABC'))
df.head(10)
```

```python
# pandas의 시각화
df.plot()
plt.title('Pandas의 내장된 plot메서드')
plt.xlabel('날짜')
plt.ylabel('Data')
```

- plot함수의 옵션
 - kind : 함수 종류 변경(다양한 프롯형태로 시각화 가능)
        bar, pie, hist, kde(밀도), box, scatter ...


```python
import seaborn as sns

iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
# 판다스의 경우 데이터 셋명을 먼저 씀
iris.sepal_length[:20].plot(kind='bar', color='y')
plt.title('iris의 꽃받침의 길이를 사용한 시각화')
plt.xlabel('index') # 기본적으로 x축이 인덱스로 지정됨
plt.ylabel('곷받침의 길이')
```

```python
# 품종을 제외한 아이리스의 컬럼을 가져오기
iris[:5].plot.bar(rot=0, color = 'cmgb')
plt.title('아이리스 붓꽃 데이터')
plt.xlabel('Index')
plt.ylabel('각 Feature의 값')
plt.ylim(0,7)
plt.legend(loc=0)
plt.show()
```

```python
# 품종별로 각 특성의 평균값을 시각화 하시오.
irisG = iris.groupby(iris.species).mean()
irisG.plot.bar(rot=0, color='bkgy')
plt.title('품종별 feature평균')
plt.xlabel('품종')
plt.ylabel('평균')
```

```python
irisG = iris.groupby(iris.species).mean()
colors = ['darkorange','yellow','goldenrod']
irisG.T.plot.bar(rot=0, color=colors) # x축 y축 변경
plt.title('품종별 feature평균')

plt.xlabel('품종')
plt.ylabel('평균')
```

```python
# 객실 등급별 승객수의 비율을 시각화
titaG = titanic.pclass.value_counts()
titaG
```

```python

titaG.plot.pie(autopct='%1.2f%%')
plt.axis('equal')

plt.show()
```

```python
iris.plot.hist(color=['tomato','springgreen','fuchsia','hotpink'])
plt.show()
```

```python
iris.plot.kde()
plt.show()
```

```python
iris.plot.box()
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=1,h_pad=2)
plt.show()
```
