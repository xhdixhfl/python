##### plot 옵션들
- color(c) : r, b, k(블랙), g, y, w,c...
- line width(lw)
- marker: 마커 크기(ms), 마커색(mec), 마커채우기(mfc),채우는 넓이(mew), s(사각형)
- 축범위: xlim, ylim
- line style(ls)
- http://matplotlib.org/examples/color/named_colors.html # 컬러
- http://matplotlib.org/api/lines_api.html#matplotlib.lines.Line2D # 속성


```python
plt.title('한글 제목')
plt.plot([10,20,30,40],[1,4,9,16], 'cs--') # 선샛, 볼릿기호, 선스타일을 한번에 넣음
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show()
```

```python
plt.plot([10,20,30,40],[1,4,9,16], c='m',
         lw=5, ls='--', marker='o', ms=15, mec='c', mew=5, mfc='k')
plt.show()
```

```python
plt.plot([10,20,30,40],[1,4,9,16], c='k',
         lw=5, ls='--', marker='o', ms=15, mec='c', mew=5, mfc='m')
# 축범위 
plt.xlim(0,50)
plt.ylim(-10,30)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
plt.plot(x,y, c='r', ls='--')
plt.yticks([-1,0,1])
plt.xticks([-np.pi,-np.pi/2, 0, np.pi/2, np.pi])
plt.show()
```

```python
# 축에 글 추가하기
x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
plt.title('연습 그래프', c='w')
plt.plot(x,y, c='r', ls='--')
plt.yticks([-1,0,1], ['low','mid','high'], c='w') # 눈금값에 문자로 대체
plt.xticks([-np.pi,-np.pi/2, 0, np.pi/2, np.pi], c='w')
plt.grid(True) # 눈금선
plt.show()
```

```python
# 여러개의 선 그래프
## 기본 잡기
t = np.arange(0., 5., 0.2)
plt.plot(t,t, 'y--', t, 0.5 * t,'rs:', t, 0.2*t**3, 'g^-')
plt.show()
```

```python
x= np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x) # 언패킹
plt.plot(x,c, 'kp', label='cosine')
plt.plot(x,s, 'b:', label='sine')
plt.legend(loc=0) # 0베스트, 1오른쪽상단, 2왼쪽상단, 3왼쪽하단, 4오른쪽하단
plt.show()
```
