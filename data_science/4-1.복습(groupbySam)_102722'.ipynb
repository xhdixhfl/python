### 그룹별 집계 (Groupby)
- groupby 메커니즘 : 분할 --> 적용 --> 결합\
- Series의 데이터 그룹화 : 특정 열인덱싱.groupby(기준열인덱싱)
- DataFrame의 데이터 그룹화 : df.groupby('기준이될 컬럼명')

```python
ipl_data = {'Team':['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings',
                    'Riders','Royals','Royals','Riders'],
           'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],
           'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
           'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df
```

```python
# 팀별 포인트의 합을 추출
#df.groupby('Points').df.sum() # try
df.groupby('Team')['Points'].sum()
df.groupby('Team').sum()['Points'] # 둘다 가능
```

```python
# 팀별, 연도별로 묶어서 Points합
multi_group = df.groupby(['Team', 'Year']).Points.sum() 
## 합하고자하는 컬럼명을 불러오는 방식은 많으니 참고
multi_group
```

```python
multi_group.index # index는 다중일때 튜플로 묶임
```

```python
# 멀티인덱스의 데빌즈와 킹즈를 뽑
multi_group.loc[['Devils','Kings']]
# 연속된 공간이라 리스트로 안묶어도 가능
multi_group.loc['Devils':'Kings'] # 문자형은 본인까지 추출
```

```python
# swaplevel : 레벨(인덱스)간의 우선순위를 변경함
# 연도와 팀의 상위인덱스 변경
multi_group.swaplevel().sort_index()
```

```python
# 레벨단위로 연산이 가능함
multi_group.sum(level=0) # 0번쨰 레벨이 팀이니깐 팀별 합계를 구해줌
```

```python
multi_group.sum(level=1) # 연도별
```

### 집계, 변환, 필터
- 집계(aggreate): 그룹화된 통계 정보를 추출할 때 사용하는 함수, agg()
- 변환(transform): 정보를 변환
- 필터(filter): 특정 정보를 제거하여 보여주는 필터링 기능

```python
# 그룹을 잡은 데이터로 진행함.
## 팀단위로 그룹화
grouped = df.groupby('Team')
grouped
```

```python
 # get_group(): 키값을 기준으로 데이터프레임 객체를 반환
grouped.get_group('Riders')
```

#### agg(): 다양한 함수를 한번에 적용하는 함수
- agg(함수명)

```python
grouped.agg(min) # 파이썬이 제공하는 함수 
```

```python
grouped.agg(np.mean)  # 넘파이가 제공하는 함수
```

```python
grouped.agg([sum, np.mean]) # 함수 두개 적용시 리스트로 묶어서 
```

#### 변환
- 그룹화 된것이 아닌 개별데이터들의 집계를 보여준다고 생각하면 됨.
- 각각의 값을 정규화 할 때(z-value, 최대, 최소 등) 단위를 맞춰주고자 할때 사용

```python
grouped.transform(max)
```

```python
z_value = lambda x : (x - x.mean()) / x.std()  # 표준화 시키는 과정
grouped.transform(z_value)
```

#### 필터
- 특정 조건에 따라 데이트럴 검색할 때 사용하는 함수

```python
# 팀 단위로 그룹바이
## 이때 팀 정보가 3개 이상인 것들만 보고 싶다
df.groupby('Team').filter(lambda x : len(x) >= 3) # len() 개수
```

```python
# 팀별일 떄, Points의 최댓값이 850보다 큰값은
df.groupby('Team')['Points'].filter(lambda x : max(x) > 850) # 시도
### 최대값을 구하고자 한 부분이 포인츠의 값이니깐 x의 포인츠의 맥스를 구한다는 거였음...
```

```python
# 샘
df.groupby('Team')['Points'].max() # Points의 최댓값 추출
df.groupby('Team').filter(lambda x : x['Points'].max() > 850)
## 추출된 값을 보면 850이상의 포인츠를 가진 팀별의 데이터만 추출된걸 볼 수 있음
```

```python
import seaborn as sns
tips = sns.load_dataset('tips') # 패키지에 저장된 내장 데이터셋 불러오기
```

```python
tips.info() # 데이터 이해 먼저
```

```python
tips.sample(5)
```

```python
# 어떤 인사이트를 도출한지 적어두면 생각하기 편함
## 많이 먹었다면 팁의 금액도 클까?? 그비율은? 어떤경우에 팁을 많이줄까?

# 청구서 금액 대비 팀의 비율
## 파생변수
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
# 수치 데이터의 기초통계량 describe()
tips.describe()
```

```python
# 성별로 그룹설정
tips.groupby('sex') # 주소만 뜸
tips.groupby('sex').count() # df형태로 출력
```

```python
tips.groupby('sex').size() # 시리즈형태로 출력
```

```python
# 성별과 흡연 여부
tips.groupby(['sex','smoker']).size() # 일반적으로 간결한 사이즈를 많이 사용
```

```python
 # 성별에 따른 평균 팁의 비율
tips.groupby('sex')['tip_pct'].mean()
```

```python
# 흡연여부에 따른 평균 팁
tips.groupby('smoker')['tip_pct'].mean()
```

```python
# 성별 흡연 여부에 따른 평균 팁비율
tips.groupby(['sex','smoker'])['tip_pct'].mean() # series
tips.groupby(['sex','smoker'])[['tip_pct']].mean() # list
```

```python
# max - min
def mmm(x):
    return x.max()- x.min() # try ㅇㅋ
```

```python
# pivot 테이블
tips.pivot_table('tip_pct', 'sex', 'smoker', aggfunc = 'count', margins = True) 
## 적용 함수, 여백설정 옵션(총합)
```

```python
tips.pivot_table('tip_pct', 'sex', 'smoker')
```

```python
tips.groupby(['sex','smoker'])[['tip_pct']].describe()
```

```python
# 각 그룹에서 가장 많은 팁과 가장 적은 팁의 차이를 확인하시오
tips.groupby(['sex','smoker'])[['tip']].agg(mmm)
```

```python
tips.groupby(['sex','smoker'])[['tip']].agg([mmm, 'mean']) # 파이썬의 mean은 ''로 묵어야 됨. 
```

```python
# 계산하는 함수에 적용컬럼을 다르게 하고 싶은경우, dict로 담는다
## agg({'colname': func})
tips.groupby(['sex','smoker']).agg({'tip': np.mean, 'total_bill' : mmm}) 
```
