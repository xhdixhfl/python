 #### 수직 막대(bar)

```python
y = [2,3,1]
x = np.arange(len(y))

plt.title('수직 막대그래프')
xlabel = ['가','나','다']
plt.bar(x,y)
plt.xticks(x, xlabel) # 눈금
plt.yticks(sorted(y)) # 높이를 뜻하는데 2,3,1 순으로 오면 안되니깐 정렬을 한것
plt.xlabel('엑스 축')
plt.ylabel('빈도수')
plt.show()
```

#### 수평막대(barh)
- 실험값에서 오차를 나타날 때 주로 사용함.

```python
np.random.seed(0) # 0보다 큰 정수면 모두 ok

people = ['몽룡','춘향','방자','향단']
yPos = np.arange(len(people)) # 막대 위치, 4개이니 막대가 4개여야함

data = np.random.rand(len(people)) * 10 # 막대 값
error = np.random.rand(len(people)) # 에러 값, 수평막대에 에러값을 표시하고자 함.

# 그래프
plt.title('수평 막대그래프')
plt.barh(yPos, data, alpha=0.4) # alpha 투명도 옵션
plt.yticks(yPos, people) # 이걸 기준으로 축에 이름을담을거야(기준, 이름)
plt.xlabel('엑스 라벨')
plt.show()
```

```python
# 그래프
plt.title('수평 막대그래프')
plt.barh(yPos, data, xerr = error) # xerr에러값 표시하는 옵션
plt.yticks(yPos, people) # 이걸 기준으로 축에 이름을담을거야(기준, 이름)
plt.xlabel('엑스 라벨')
plt.show()
```

### 파이 차트(pie chart)
- 카테고리별 값의 상대적인 비교를 해야할 떄 주로 사용한다.
- 파이차트를 그릴 떄 원의 형태를 유지할 수 있도록 plt.asix('equal')명령을 담아야함.
 -http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie

```python
labels = ['맥주','소주','양주','막걸리']
sizes = [45,12,18,25]
colors = ['gold','cyan','brown', 'magenta']
explode = (0.05, 0, 0, 0) # 특정 파이만 뺴내오기
plt.title('주종 인기도')
plt.pie(sizes,  explode = explode,labels =labels, 
        colors = colors, autopct = '%1.1f%%', shadow=True, startangle=90) 
## %는 하나는 서식기호로 인식해서 두개해야 문자로 인식 1.1(소숫점 한자리)
plt.axis('equal')
plt.show()
```

```python
labels = ['맥주','소주','양주','막걸리']
sizes = [45,12,18,25]
colors = ['red','green','yellow', 'blue']
explode = (0.05, 0, 0, 0)
wedgeprops={'width': 0.7, 'linewidth': 5} # 도넛 파이 옵션
plt.title('주종 인기도')
plt.pie(sizes,  explode = explode,labels =labels, 
        colors = colors, autopct = '%1.1f%%', shadow=True, startangle=90,wedgeprops = wedgeprops) 
## %는 하나는 서식기호로 인식해서 두개해야 문자로 인식 1.1(소숫점 한자리)
plt.axis('equal')
plt.show()
```

#### 버블차트(bubble chart)
- 2차원이 아니라 3차원 또는 4차원의 산점도를 표현할 경우 점 하나의 크기 혹은 색을 이용하여 다른 데이터값을 표시한다.
 - s=크기, c=색깔...
- 2차원 그래프 밖에 그리지 못하기때문에 다른 차원은 크기또는 색으로 구별을 준다는거

```python
N = 30 # 데이터 개수(전체)
np.random.seed(1)

x=np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15*np.random.rand(N)) ** 2

plt.title('bubble chart')
plt.scatter(x, y1, c = y2, s = y3)
plt.show()
```

### Imshow() <머신러닝에서 할거 맛보기>
- 화상(image)데이터를 표현하는 시각화 도구
- 이미지데이터도 행과 열을 가진 2차원 데이터(3,4차원도 가능)
-http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.imshow


```python
# 필요 도구 로딩
## 손으로 그린 숫자 데이터를 이용
from sklearn.datasets import load_digits
digits = load_digits()
digits
```

```python
x = digits.images[1]
x # 하나의 숫자를 만들어내는 좌표값
```

```python
plt.title('손글씨데이터셋(mnist digits:0)')
plt.imshow(x, interpolation='spline16', cmap=plt.cm.Pastel1)
plt.xticks([])
plt.yticks([])
plt.grid(False)

plt.show()
```

```python
plt.title('손글씨데이터셋(mnist digits:0)')
plt.imshow(x, interpolation='nearest', cmap=plt.cm.spring_r)

plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.show()
```

```python
methods = ['none', 'nearest', 'bilinear', 'bicubic', 'spline16', 
          'spline36', 'hanning','hamming', 'hermite','kaiser', 'quadric', 
          'catrom', 'gaussian', 'bessel','mitchell', 'sinc', 'lanczos']

# 3행 6열로 분할 할건데 전체 도화지 사이즈는 12,6 짜리이다. 
fig, axes = plt.subplots(3,6, figsize=(12,6),
                        subplot_kw={'xticks':[], 'yticks':[]})

# 메서드 만큼 반복처리해서 시각화 할거다, zip은 변수 묶는것
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(x, interpolation=interp_method, cmap=plt.cm.PiYG)
    ax.set_title(interp_method)
    
plt.show()
```

#### distplot
- 1차원 실수의 분포 플롯
- rugplot은 데이터 위치를 x축 위의 작은 선분(rug)로 나타 내어 실제 데이터의 위치를 보여준다.

```python
x = iris.petal_width.values # 아이리스의 페탈넓이 컬럼의 값만 가져오자
sns.kdeplot(data=iris, x ='petal_width')
sns.rugplot(x) # 데이터 위치
plt.winter()
plt.show()
```

```python
# 속성 담기
sns.distplot(x, kde=True, rug=True, color = 'r')
plt.title('넣어보wk')
plt.show()
```

```python
pg = sns.load_dataset('penguins')
```

```python
sns.displot(data=pg, x='flipper_length_mm', hue='species',
            multiple='stack', kind='kde', rug=True, palette='spring')
# multiple='stack' 모듈내 기본코드를 공유하고 다른구성요소가 없는 경우 유사한 기능을 제공하는 옵션
plt.show()
```
