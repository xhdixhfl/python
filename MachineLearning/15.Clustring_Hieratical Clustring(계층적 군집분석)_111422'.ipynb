# Hieratical Clustring
- 계층적 트리 모형을 이용해 개별 개체들을 순차적, 계층적으로 유사한 개체 내지 그룹과 통합하여 군집화를 수행하는 알고리즘
- K-means 군집화와 달리 <span style= 'color:dodgerblue'>**군집수를 사전에 정하지 않아도 됨(Dendogram 이용)**</span>
- 거리 또는 유사도(유클리드 거리 계산법)를 기반으로 클러스터를 형성하는 알고리즘

## Agglomerative Method : 응집형 계층적 클러스터링
- 각 데이터 지점에서 시작해 유사한 지점들을 함께 결합해 클러스터를 형성(Bottom-Up 방식)
- 연결 기준에 따라 군집이 달라지므로 <span style= 'color:dodgerblue'>데이터 특성, 분포에 따라 적절히 변경</span>하며 확인해야 한다.
- <span style= 'color:dodgerblue'>**연결기준(유클리디안 거리)**</span>
 1. **Single Linkage**: 두 클러스터 내의 가장 가까운 점 사이의 거리
  > 먼 거리에 있는 이상한 데이터의 영향을 최소화 할 수 있다.
 2. **Complete Linkage** : 두 클러스터 내의 가장 먼 점 사이의 거리
  > 클러스터간 거리가 멀고 클러스터의 밀집도가 높을 경우 좋은 성과를 낸다.
 3. A**verage Linkage** : 두 클러스터 내의 모든 점 사이의 평균 거리
  > 무난한 방법이나 데이터에 노이즈가 많고 중심점이 명확하지 않으면 잘 동작하지 않을 수 있다

## Dendrogram 시각화
- 적절한 k값을 찾는 방법

```python
from scipy.cluster.hierarchy import linkage, dendrogram 
# linkage(t= feature, method='연결방식', metric='거리계산방식')
```

```python
# 연결방식을 각각 다르게 하는 3개의 변수 생성
clusters_single = linkage(y=data, method='single', metric='euclidean')
clusters_average = linkage(y=data, method='average', metric='euclidean')
clusters_complete = linkage(y=data, method='complete', metric='euclidean')
```

```python
# 위에서 정의한 3변수의 시각화
fig, axes = plt.subplots(3,1) # 3행1열의 그래프를 그릴거얌
fig.set_size_inches(15,15) # inch단위로 그래프 크기 재설정

ax1 = dendrogram(clusters_single, leaf_font_size=0, ax=axes[0]) # 그림을 더 크게 보려구 0
ax2 = dendrogram(clusters_average, leaf_font_size=0, ax=axes[1])
ax3 = dendrogram(clusters_complete, leaf_font_size=0, ax=axes[2])
```

- single(노이즈가 심함)은 데이터가 왼편으로 너무 치우쳐서 쓰기 힘들듯하니 avg, com를 써야겠군
- k는 avg이면 3개, com이면 4개를 쓰는게 가장 적절할 듯
- complete가 일반적으로 가장 많이 사용됨

## Data preprocessing _ scaler, pca(차원축소)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 스케일링
scaler = StandardScaler()
df.loc[:,:] = scaler.fit_transform(data)
df.describe().loc[['mean','std']]
```

```python
# 차원축소
pca = PCA(n_components = 2) # 0~1사이의 값이 나오면 신뢰도(0.95,0.99)가 나옴, 그렇지 않으면 차원수
```

```python
data = pca.fit_transform(data)
data.shape # 축소된 거 확인
```

## Agglomerative Clustring
### single linkage

```python
from sklearn.cluster import AgglomerativeClustering
single_clustering = AgglomerativeClustering(n_clusters=3, linkage='single')
single_clusters = single_clustering.fit_predict(data)
single_clusters[:10]
```

```python
temp = pd.DataFrame(data, columns=['pca1','pca2'])
temp['single'] = single_clusters
```

```python
# single linkage 시각화
sns.lmplot(x='pca1', y='pca2',data=temp, hue='single', fit_reg=False, palette='twilight') # 회귀선 안보이게
```

### average linkage

```python
average_clustering = AgglomerativeClustering(n_clusters=3, linkage='average')
average_clusters = average_clustering.fit_predict(data)
temp['average'] = average_clusters
```

```python
# average linkage 시각화
sns.lmplot(x='pca1', y='pca2',data=temp, hue='average', fit_reg=False, palette='twilight')
```

### complete linkage

```python
complete_clustering = AgglomerativeClustering(n_clusters=3, linkage='complete')
complete_clusters = complete_clustering.fit_predict(data)
temp['complete'] = complete_clusters
```

```python
# complete linkage 시각화
sns.lmplot(x='pca1', y='pca2',data=temp, hue='complete', fit_reg=False, palette='twilight')
```

- 시각화 한 결과 ave의 반영이 가장 좋을것으로 보임... 그럼 군집수가 4 또는 5라면?

```python
raverage_clustering = AgglomerativeClustering(n_clusters=5, linkage='average')
raverage_clusters = average_clustering.fit_predict(data)
temp['raverage'] = average_clusters
```

```python
sns.lmplot(x='pca1', y='pca2',data=temp, hue='raverage', fit_reg=False, palette='twilight')
```
