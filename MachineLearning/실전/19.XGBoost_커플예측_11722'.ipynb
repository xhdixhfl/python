## 피처 엔지니어링
<br>

### 나이gap(age, age_o)
- 매칭 기준 연상연하 판단한 변수
- 남자의 나이를 기준으로 본인이 남자면(본인 나이 - 상대 나이), 상대가 남자면(상대 나이 - 본인)
- 무응답 -99로 넣은게 있으니 둘중 한 명이라도 -99를 뛰면 무응답이 출력되도록
- 남자가 연상이면 +, 여자가 연상이면 -

```python
data['gender']
```

```python
def age_gap(x):
    if x['age'] == -99: # 결측처리의 경우
        return -99
    elif x['age_o'] == -99:
        return -99
    elif x['gender'] == 'female':
        return x['age_o'] - x['age']
    else:
        return x['age'] - x['age_o']
```

```python
data['age_gap'] = data.apply(age_gap, axis = 1) # 열단위
```

### 인종(race, race_o)
- 나와 상대가 같은 인종이면(1), 그렇지 않으면(-1), 결측값의 경우(-99)

```python
def same_race(x):
    if x['race'] == -99: # 결측처리의 경우
        return -99
    elif x['race_o'] == -99:
        return -99
    elif x['race'] == x['race_o']:
        return 1
    else:
        return -1
```

```python
data['same_race'] = data.apply(same_race, axis = 1)
```

```python
# 같은 인종의 중요성
data['importance_same_race'].unique()
```

```python
# 중요 특성에 가중치를 부여하기 위한 함수 정의
def same_race_point(x):
    if x['same_race'] == -99:
        return -99
    else:
        return x['same_race'] * x['importance_same_race']
```

```python
data['race_point'] = data.apply(same_race_point, axis = 1)
```

### 평가항목들(6개*4set) 점수 환산
- 상대방이 중요하게 여기는 정도, 나를 평가한점수
- 내가 중요하게 여기는 정도, 상대방을 평가한 점수 
- 평가점수를 바탕으로 중요도를 가중치 처리하여 점수로 환산
- 내가평가한점수, 상대방이 평가한 점수 이렇게 2개로 생성 

```python
# 점수 환산을 위한 함수 정의
def rating(data, importance, score):
    if data[importance] == -99: # 결측치
        return -99
    elif data[score] == -99:
        return -99
    else:  # 가중치 부여하여 점수 환산
        return data[importance] * data[score]
```
