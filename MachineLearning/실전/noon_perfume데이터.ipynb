```python
# gender 수치화 하는 작업.
for x in range(len(data)):
    if data.department[x] == 'Men':
        data['gender'][x] = 0
    elif data.department[x] == 'Women':
        data['gender'][x] = 1
    elif data.department[x] == 'Unisex':
        data['gender'][x] = 2
    else:
        data['gender'][x] = 3
```

```python
# 가격을 ml당으로 환산해서
data['price/ml'] = (data['new_price']/data['ml']).round(3)
data.head()
```

```python
# 성졓
men = data[data['department']=='Men']
women = data[data['department']=='Women']
kids = data[data['department']=='Kids unisex']
unisex = data[data['department']=='Unisex']
```

```python
f, ax = plt.subplots(figsize = (10,5))
plt.hist(men['price/ml'], alpha=0.4, label = 'Men', density=False, color='c')
plt.hist(women['price/ml'], alpha=0.4, label = 'Women', density=False, color='m')
plt.hist(unisex['price/ml'], alpha=0.4, label = 'Unisex', density=False, color='y')

plt.yscale('log')
plt.xticks(ticks=[0,2,4,6,8,10,12], labels=None)
plt.title('Comparsion of Price per ml')
plt.xlabel('Price per ml')
plt.ylabel('Number of Items')
plt.legend()
plt.show()
```

```python
rate_jitter = data['item_rating'] + np.random.normal(0,0.1, size=len(data))
sns.relplot(x=data['price/ml'], y=rate_jitter, kind='scatter',
            hue=data['department'], alpha=0.2, size=data['price/ml'],
           col = data['department'], col_wrap=2)
plt.xticks(ticks=[0,2,4,6,8,10,12], labels=None)
plt.show()
```

```python
from scipy.stats import linregress
regression = linregress(men['price/ml', men['item_rating']])

fx = np.array([men['price/ml'].min(), men['price/ml'].max()])
fy = regression.intercept + regression.slope*fx # 방정식
rate_jitter_men = men['item_rating']+np.random.normal(0,0.1, size=len(men))
sns.replot(x=men['price/ml'], y=rate_jitter_men, kind='scatter', 
           alpha=0.2, size=men['price/ml'])
plt.xticks(ticks=[0,2,4,6,8,10,12], labels=None)
plt.plot(fx, fy, '--', color='red')
plt.xscale("log")
plt.title('Regression between the Price and the Stars', y=1.04)
plt.xlabel('Price per ml')
plt.ylabel('Stars')
plt.show()
```
