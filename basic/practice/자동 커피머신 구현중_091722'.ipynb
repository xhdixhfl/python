     커피 자동 판매기 프로그램을 구현하시오.
     커피 판매기에는 커피 종류별 재료가 아래와 같이 셋팅 되어있다.
     커피3, 물 3, 설탕 3, 프림3, 컵3
     블랙커피 = 커피1 + 물1 + 컵1
     설탕커피 = 커피1 + 물1+ 설탕1 + 컵1
     프림커피 = 커피1 + 물1 + 프림1 + 컵1
     밀크커피 = 커피1 + 물1 + 설탕1 + 프림1 + 컵1
     재료 부족시 판매불가, 재료 보충시 모든재료가 3으로 설정되거나 필요한 재료만 3으로 셋팅하는 것도 가능
     각 자료당 3개씩 셋팅가능, 커피가격은 블랙,설탕(1000), 프림(1500), 밀크(2000)

```python
def showMenu():
    print("="*35)
    print("====== 커피 자동 판매기 ======")
    print("=" * 35)
    print("1. 블랙커피(1000원)")
    print("2. 설탕커피(1000원)")
    print("3. 프림커피(1500원)")
    print("4. 밀크커피(2000원)")
    print("-"*35)

    #print(input("주문번호 금액 = "))

def subMenu():
    print(" ====== 자동판매기 관리 프로그램 ======")
    print("[1] 재료보충  [2] 금액조정  [3]작업종료")
    print("=" * 25)
    #print(input("메뉴 = "))

def plusMenu():
    print("          ====== 재료보충 ======         ")
    print("[1]전체 [2]커피 [3]물 [4]설탕 [5]프림 [6] 컵")
    print("=" * 35)
    #print(input("선택 = "))

def priceMenu():
    print("          ========   금액조정   ========         ")
    print("[1]전체 [2]블랙커피 [3]설탕커피 [4]프림커피 [5]밀크커피")
    print("=" * 35)
    #print(input("선택 = "))
                   
```

```python
# 수량과 가격조정을 위한 리스트들
black = ['블랙커피', 1000, 3]
sugar = ['설탕커피', 1000, 3]
prim = ['프림커피', 1500, 3]
milk = ['밀크커피', 2000, 3]  # 밀크가 물 수량이라고 생각
cup = ['종이컵', 0, 3]

# 수량 상승 함수
def plusB():
    black[-1] = 3

def plusS():
    sugar[-1] = 3
    
def plusP():
    prim[-1] = 3
    
def plusM():
    milk[-1] = 3
    
def plusC():
    cup[-1] = 3
    
def plusA():
    black[-1] = 3
    sugar[-1] = 3
    prim[-1] = 3
    milk[-1] = 3
    cup[-1] = 3
    
# 수량 감소 함수
def minusB():
    black[-1] = black[-1] - 1

def minusS():
    sugar[-1] = sugar[-1] - 1
    
def minusP():
    prim[-1] = prim[-1] - 1
    
def minusM():
    milk[-1] = milk[-1] - 1
    
def minusC():
    cup[-1] = cup[-1] - 1
    
    
# 가격 조정 함수
def priceB():
    black[1] = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))

def priceS():
    sugar[1] = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))
    
def priceP():
    prim[1] = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))
    
def priceM():
    milk[1] = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))
    
# def priceC():
#     cup[1] = int(input("조정하고자 하는 가격을 숫자만 입력해주세요 :"))
    
def priceA():
    a = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))
    black[1] = a
    sugar[1] = a
    prim[1] = a
    milk[1] = a
```

```python
def coffeeMachine():
    while True:
        showMenu()
        price = input("투입금액 = ")
        if price == "Manager":
            subMenu()
            while True:
                
                menu = input("메뉴 = ")
                if menu == "1":
                    plusMenu()
                    mn1 = input("선택 = ")
                    if mn1 == "1":
                        plusA()
                        print("전체 수량을 3개씩 충전하였습니다.")
                        return coffeeMachine()
                    elif mn1 == "2":
                        plusB()
                        print("커피원두의 수량을 3개씩 충전하였습니다.")
                        return coffeeMachine()
                    elif mn1 == "3":
                        plusM()
                        print("물의 수량을 3개씩 충전하였습니다.")
                        return coffeeMachine()
                    elif mn1 == "4":
                        plusS()
                        print("설탕의 수량을 3개씩 충전하였습니다.")
                        return coffeeMachine()
                    elif mn1 == "5":
                        plusP()
                        print("프림의 수량을 3개씩 충전하였습니다.")
                        return coffeeMachine()
                    elif mn1 == "6":
                        plusC()
                        print("컵의 수량을 3개씩 충전하였습니다.")
                        return coffeeMachine()
                    else:
                        print("올바르지 않는 선택입니다. 다시 선택해주세요.")
                        return subMenu()                
                elif menu == "2":
                    priceMenu()
                    mn2 = input("선택 = ")
                    if mn2 == "1":
                        priceA()
                    elif mn2 == "2":
                        priceB()
                    elif mn2 == "3":
                        priceS()
                    elif mn2 == "4":
                        priceP()
                    elif mn2 == "5":
                        priceM()
                    else:
                        print("올바르지 않는 선택입니다. 다시 선택해주세요.")
                        return subMenu()

                elif menu == "3":
                    print("\t재료보충 프로그램을 종료합니다.")
                    break
                    return coffeeMachine()
                else:
                    print("잘못 입력하셨습니다. 다시 선택해 주세요")
                    return subMenu()
            
        else:
            global choice
            choice = input("주문번호 금액 = ")
        if choice == "1":
            if int(price) >= black[1]:
                if black[-1] >= 1 and milk[-1] >= 1 and cup[-1] >= 1:
                    minusB()
                    minusM()
                    minusC()
                    print("블랙커피가 나왔습니다.", "잔돈 = %d"%(int(price)- black[1]))
                    break
                else:
                    print("해당커피의 수량이 부족합니다. 다른 커피를 이용해주세요.")
                    return coffeeMachine()
            else:
                print("투입금액이 부족합니다. 다른 커피를 이용해주세요.")
                return coffeeMachine()
                          
        elif choice == "2":             
            if int(price) >= sugar[1]:
                if black[-1] >= 1 and milk[-1] >= 1 and sugar[-1] >= 1 and cup[-1] >= 1:
                    minusB()
                    minusM()
                    minusS()
                    minusC()
                    print("설탕커피가 나왔습니다.", "잔돈 = %d"%(int(price)- sugar[1]))
                    break
                else:
                    print("해당커피의 수량이 부족합니다. 다른 커피를 이용해주세요.")
                    return coffeeMachine()
            else:
                print("투입금액이 부족합니다. 다른 커피를 이용해주세요.")
                return coffeeMachine()
                          
        elif choice == "3":
            if int(price) >= prim[1]:
                if black[-1] >= 1 and milk[-1] >= 1 and prim[-1] >= 1 and cup[-1] >= 1:
                    minusB()
                    minusM()
                    minusP()
                    minusC()
                    print("설탕커피가 나왔습니다.", "잔돈 = %d"%(int(price)- prim[1]))
                    break
                else:
                    print("해당커피의 수량이 부족합니다. 다른 커피를 이용해주세요.")
                    return coffeeMachine()
            else:
                print("투입금액이 부족합니다. 다른 커피를 이용해주세요.")
                return coffeeMachine()
                          
        elif choice == "4":
            if int(price) >= milk[1]:
                if black[-1] >= 1 and milk[-1] >= 1 and sugar[-1] >= 1 and prim[-1] >= 1 and cup[-1] >= 1:
                    minusB()
                    minusM()
                    minusS()
                    minusP()
                    minusC()
                    print("설탕커피가 나왔습니다.", "잔돈 = %d"%(int(price)- milk[1]))
                    break
                else:
                    print("해당커피의 수량이 부족합니다. 다른 커피를 이용해주세요.")
                    return coffeeMachine()
            else:
                print("투입금액이 부족합니다. 다른 커피를 이용해주세요.")
                return coffeeMachine()
        else:
            print("잘못 입력하셨습니다. 다시 선택해주세요")  # 만약 Manager 왜 문자가 들어오면 어떻게 하지??
```
