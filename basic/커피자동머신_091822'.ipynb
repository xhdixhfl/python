```python
def showMenu():  # 전체 메뉴판
    print("="*35)
    print("====== 커피 자동 판매기 ======")
    print("=" * 35)
    print("1. 블랙커피(1000원)")
    print("2. 설탕커피(1000원)")
    print("3. 프림커피(1500원)")
    print("4. 밀크커피(2000원)")
    print("*종료를 원하시면 투입 금액에 exit를 입력하세요.*")
    print("-"*35)

    #print(input("주문번호 금액 = "))

def subMenu(): # Manager 입력시 출력되는 메뉴판
    print(" ====== 자동판매기 관리 프로그램 ======")
    print("[1] 재료보충  [2] 금액조정  [3]작업종료")
    print("=" * 25)
    while True:
        menu = input("메뉴 = ")
        if menu == "1":
            plusMenu()
            mn1 = input("선택 = ")
            if mn1 == "1":
                plusA()
                print("전체 수량을 3으로 충전하였습니다.")
                return coffeeMachine()
            elif mn1 == "2":
                plusB()
                print("커피원두의 수량을 3으로 충전하였습니다.")
                return coffeeMachine()
            elif mn1 == "3":
                plusM()
                print("물의 수량을 3으로 충전하였습니다.")
                return coffeeMachine()
            elif mn1 == "4":
                plusS()
                print("설탕의 수량을 3으로 충전하였습니다.")
                return coffeeMachine()
            elif mn1 == "5":
                plusP()
                print("프림의 수량을 3으로 충전하였습니다.")
                return coffeeMachine()
            elif mn1 == "6":
                plusC()
                print("컵의 수량을 3으로 충전하였습니다.")
                return coffeeMachine()
            else:
                print("올바르지 않는 선택입니다. 다시 선택해주세요.")
                return subMenu()                
        elif menu == "2":
            priceMenu()
            mn2 = input("선택 = ")
            if mn2 == "1":
                priceA()
                print("전체 가격 변경이 완료되었습니다.")
                return coffeeMachine()
            elif mn2 == "2":
                priceB()
                print("블랙커피의 가격 변경이 완료되었습니다.")
                return coffeeMachine()
            elif mn2 == "3":
                priceS()
                print("설탕커피의 가격 변경이 완료되었습니다.")
                return coffeeMachine()
            elif mn2 == "4":
                priceP()
                print("프림커피의 가격 변경이 완료되었습니다.")
                return coffeeMachine()
            elif mn2 == "5":
                priceM()
                print("밀크커피의 가격 변경이 완료되었습니다.")
                return coffeeMachine()
            else:
                print("올바르지 않는 선택입니다. 다시 선택해주세요.")
                return subMenu()
        
        elif menu == "3":
            print("\t재료보충 프로그램을 종료합니다.")
            break
            return coffeeMachine()
        else:
            print("잘못 입력하셨습니다. 다시 선택해 주세요")
            return subMenu()

def plusMenu(): # 재류보충 선택시 출력되는 메뉴판
    print("          ====== 재료보충 ======         ")
    print("[1]전체 [2]커피 [3]물 [4]설탕 [5]프림 [6] 컵")
    print("=" * 35)
    #print(input("선택 = "))

def priceMenu():   # 금액조정 선택시 출력되는 메뉴판
    print("          ========   금액조정   ========         ")
    print("[1]전체 [2]블랙커피 [3]설탕커피 [4]프림커피 [5]밀크커피")
    print("=" * 35)
    #print(input("선택 = "))
```

```python
# 수량과 가격조정을 위한 리스트들
black = ['블랙커피', 1000, 3]
sugar = ['설탕커피', 1000, 3]
prim = ['프림커피', 1500, 3]
milk = ['밀크커피', 2000, 3]  # 밀크가 물 수량이라고 생각
cup = ['종이컵', 0, 3]

# 수량 상승 함수
def plusB():  # 커피 재고
    black[-1] = 3

def plusS():  # 설탕 재고
    sugar[-1] = 3
    
def plusP():  # 프림 재고
    prim[-1] = 3
    
def plusM():  # 물 재고
    milk[-1] = 3
    
def plusC():  # 컵 재고 
    cup[-1] = 3
    
def plusA():  # 전체 재고
    black[-1] = 3
    sugar[-1] = 3
    prim[-1] = 3
    milk[-1] = 3
    cup[-1] = 3
    
# 수량 감소 함수
def minusB():
    black[-1] = black[-1] - 1

def minusS():
    sugar[-1] = sugar[-1] - 1
    
def minusP():
    prim[-1] = prim[-1] - 1
    
def minusM():
    milk[-1] = milk[-1] - 1
    
def minusC():
    cup[-1] = cup[-1] - 1
    
    
# 가격 조정 함수
def priceB():   # 블랙 커피 가격
    black[1] = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))

def priceS():   # 설탕 커피 가격
    sugar[1] = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))
    
def priceP():  # 프림 커피 가격
    prim[1] = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))
    
def priceM():  # 밀크 커피 가격
    milk[1] = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))

def priceA():   # 전체 커피 가격
    a = int(input("조정하고자 하는 가격으로 숫자만 입력해주세요 :"))
    black[1] = a
    sugar[1] = a
    prim[1] = a
    milk[1] = a
```

```python
def coffeeMachine():  # 머신 구동 함수
    while True:
        showMenu()
        global price
        price = input("투입금액 = ")
        if price == "Manager":
            subMenu()
        if price == "exit":
            print("커피자동머신 프로그램을 종료하겠습니다.")
            #break
            return
        else:
            sellCoffee()
```
