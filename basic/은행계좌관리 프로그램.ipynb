```python
# Entity Class : 단순히 데이터를 관리할 목적으로 구현하는 클래스
# 한 명의 고객에 대한 정보를 처리하는 클래스
class Account:
    
    # 계좌를 개설할 때
    def __init__(self, num, name, balance):
        self.num = num
        self.name = name
        self.balance = balance
        
    # 입금처리
    def inMoney(self, money):
        self.balance += money
        
    # 출금처리
    def outMoney(self, money):
        if money > self.balance:
            return 0
        else:
            self.balance -= money
            return money
        
    # 계좌번호 반환
    def getNum(self):
        return num
    
    # 고객정보 출력
    def disp(self):
        print(self.num, self.name, self.balance, sep = "\t")
        
    # 잔액 반환
    def getBalance(self):
        return self.balance
```

```python
# 여러 명의 계좌정보를 저장하기 위한 기억공간
accList = {}  # accList = dict()로 구현가능

# 메뉴 출력하는 메서드
def showMenu():
    print("\n == Bank Menu ==")
    print("1. 계좌개설")
    print("2. 입금처리")
    print("3. 출금처리")
    print("4. 전체고객 잔액조회")
    print("5. 프로그램 종료")
    print("=" * 30)
 

 # 계좌 개설을 위한 메서드
def makeAccount():
    num = input("계좌번호 = ")
    
    while num in accList.keys():
        num = input("계좌번호 = ")
        
    name = input("고객이름 = ")
    money = int(input("예금금액 = "))
    acc = Account(num, name, money)
    accList[num] = acc
    print("계좌 개설이 완료되었습니다.")
    
# 입금처리를 위한 메서드
def withDraw():
    num = input("계좌번호 = ")
    
    while num in accList.keys():
        money = int(input("예금금액 = "))
        accList[num].inMoney(money)
        break
    else:
        print("계좌번호가 존재하지 않습니다.")
        
# 출금처리를 위한 메서드
def deposit():
    num = input("계좌번호 = ")
    
    while num in accList.keys():
        money = int(input("출금금액 = "))
        if accList[num].outMoney(money) > 0:
            print("출금처리가 완료되었습니다.")            
        else:
            print("계좌의 잔액이 부족합니다.")
        break
    else:
        print("계좌번호가 존재하지 않습니다.")
        
# 전체 고객에 대한 잔액을 조회하는 메서드
def showAllAccount():
    print("\n == 전체고객 잔액조회 ==")
    for i in accList.keys():
        accList[i].disp()
```

```python
print("은행 계좌관리 프로그램을 시작합니다.")

while True:
    showMenu()
    menu = int(input("menu = "))
    
    if menu == 1:
        makeAccount()
    elif menu == 2:
        withDraw()
    elif menu == 3:
        deposit()
    elif menu == 4:
        showAllAccount()
    elif menu == 5:
        break
    else:
        print("메뉴의 선택이 올바르지 않습니다.")
        
print("은행 계좌관리 프로그램을 종료합니다.")
```

```python
# 고객 클래스(ACCOUNT)
class customer:
    number = set()
    name = ''
    balance = int()
    
    def __init__(self, number, name, balance):
        self.number = number
        self.name = name
        self.balance = balance
    
    def put_m(self, money):
        self.balance += money
        
    def out_m(self, money):
        if money > self.balance:
            print("잔액이 부족합니다.")
            return 0
        else:
            self.balance -= money
            
    
    def getNum(self, number):
        return self.number
    
    
    def getB(self, number):
        return self.balance
    
    def getBalance(self):
        return self.balance
    
    def disp(self):
        print(self.number)
        print(self.name)
        print(self.balance)
```

```python
# 메뉴판
def menu_program():
    print("== 계좌관리 프로그램 ==")
    print("1. 계좌개설")
    print("2. 입금처리")
    print("3. 출금처리")
    print("4. 전체고객 잔액조회")
    print("5. 프로그램 종료")
    print("="*20)
```

```python
# 입금처리
def put_money():
        num = int(input('계좌입력 = '))
        for account in accountList:
            if num == account.getNum(num):
                print("계좌번호 =", account.getNum(num))
                money = int(input("입금금액 = "))
                account.put_m(money)
                print("예금잔액 =", account.getB(num))
                print("="*15)
                print("입금이 완료되었습니다.")
            else:
                print("계좌번호가 존재하지 않습니다.")
```

```python
# 출금처리
def out_money():
        num = int(input('계좌입력 = '))
        for account in accountList:
            if num == account.getNum(num):
                
                print("계좌번호 =", account.getNum(num))
                money = int(input("입금금액 = "))
                account.out_m(money)
                
                print("예금잔액 =", account.getB(num))
                print("="*15)
                print("출금이 완료되었습니다.")
            else:
                print("계좌번호가 존재하지 않습니다.")
```

```python
# 전체 예금액조회
def checkAll():
        print("\n== 전체고객 잔액조회 ==")
        print("계좌번호", " 고객이름", " 예금잔액")
        print('-'*15)
        for account in accountList:
            account.disp()
        print('-'*15)
        
        tot = 0
        for account in accountList:
            tot += account.getBalance()
        print("총 보유예금액:", tot)
```

```python
# 프로그램
accountList = []
unique = set()      ## 계좌번호 검증정보 추가


while True:
    menu_program()
    
    menu = int(input("menu = "))
    
    if menu == 1:            
        number = int(input('계좌번호 = '))

        
        if number not in unique:
            name = input('고객이름 = ')
            balance = int(input('예금금액 = '))
            account = customer(number, name, balance)
        
            accountList.append(account)
            print("계좌등록이 완료되었습니다.")
            
            unique.add(number)
            
        else:                    
            print("계좌가 중복됩니다.")
        
        
    
    elif menu == 2:
        put_money()
        
        
    elif menu == 3:
        out_money()
        
    elif menu == 4:
        checkAll()
        
    elif menu == 5:
        print("프로그램을 종료합니다.")
        break
    else:
        print("메뉴의 선택이 올바르지 않습니다.")
```
